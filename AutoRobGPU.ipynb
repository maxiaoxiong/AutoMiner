```python

# 获取当前用户设备列表
url = "https://console.vast.ai/api/v0/instances?api_key=" + api_key

payload = {}
headers = {
    'Accept': 'application/json'
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```
